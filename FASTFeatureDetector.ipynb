```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
img = cv.imread('qxy.jpg')
```

```python
# initial FAST object with default values
fast = cv.FastFeatureDetector_create()
```

```python
# find and draw the keypoints
kp = fast.detect(img,None)
img2 = cv.drawKeypoints(img, kp, None, color=(255,0,0))
```

```python
# print all default params
print("Threshold:{}".format(fast.getThreshold()))
print("Non-Max-Suppression:{}".format(fast.getNonmaxSuppression()))
print("Neighborhood:{}".format(fast.getType()))
print("Total Keypoints with nonmaxSuppression:{}".format(len(kp)))
```

```python
cv.imwrite('fast_true.png',img2)
```

```python
plt.subplot(121),plt.imshow(img,cmap='gray'),plt.title('origin'),plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(img2,cmap='gray'),plt.title('fast'),plt.xticks([]),plt.yticks([])
plt.show()
```

```python
# disable nonmaxsuppression
fast.setNonmaxSuppression(0)
kp = fast.detect(img,None)
```

```python
print("Total Keypoints without nonmaxSuppression: {}".format(len(kp)))
```

```python
img3 = cv.drawKeypoints(img,kp,None,color=(255,0,0))
cv.imwrite('fast_false.png',img3)
```

```python
plt.subplot(131),plt.imshow(img,cmap='gray'),plt.title('origin'),plt.xticks([]),plt.yticks([])
plt.subplot(132),plt.imshow(img2,cmap='gray'),plt.title('fast_true'),plt.xticks([]),plt.yticks([])
plt.subplot(133),plt.imshow(img3,cmap='gray'),plt.title('fast_false'),plt.xticks([]),plt.yticks([])
plt.show()
```

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('qxy.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
corner = cv2.goodFeaturesToTrack(gray, 25, 0.01, 10)
corners = np.int0(corner)
```

```python
for i in corners:
    x, y = i.ravel()
    cv2.circle(img, (x,y), 3, 255, -1)
```

```python
plt.imshow(img)
plt.show()
```

```python
import numpy as np
import cv2
img = cv2.imread('qxy.jpg',0)
rows, cols = img.shape
M = np.float32([[1, 0, 100],[0,1,50]])
dst = cv2.warpAffine(img, M, (rows, cols))
cv2.imshow('img', dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import os
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
img = cv.imread('qxy.jpg',0)
rows,cols = img.shape
M = np.float32([[1,0,100],[0,1,50]])
dst = cv.warpAffine(img,M,(cols,rows))
plt.imshow(dst, cmap = 'gray', interpolation = 'bicubic')
plt.show()
#cv.waitKey(0)
#cv.destroyAllWindows()
```

```python
import os
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
img = cv.imread('qxy.jpg',0)
rows,cols = img.shape
M = cv.getRotationMatrix2D((cols/2,rows/2), 60,1)# cv.getRotationMatrix2D((center_x,center_y), jiaodu,bieshu)
dst = cv2.warpAffine(img,M,(cols,rows))
plt.imshow(dst, cmap = 'gray', interpolation = 'bicubic')
plt.show()
```

```python
import os
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
img = cv.imread('qxy.jpg')
rows,cols,ch = img.shape
pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])
M = cv.getAffineTransform(pts1,pts2)
dst = cv2.warpAffine(img,M,(cols,rows))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

```python
import os
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
img = cv.imread('qxy.jpg')
rows,cols,ch = img.shape
pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])

M = cv.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(300,250))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

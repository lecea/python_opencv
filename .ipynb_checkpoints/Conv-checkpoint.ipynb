```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
img = cv.imread('qxy.jpg')

kernel = np.ones((5,5),np.float32)/25
dst = cv.filter2D(img,-1,kernel)
```

```python
plt.subplot(121)
plt.imshow(img)
plt.title('original')
plt.xticks([])
plt.yticks([])
plt.subplot(122)
plt.imshow(dst)
plt.title('averaging')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
img = cv.imread('median.jpg')
# average
blur1 = cv.blur(img,(5,5))
# gaussian
blur2 = cv.GaussianBlur(img,(5,5),0)
# median blurring
blur3 = cv.medianBlur(img,5)
# Bilateral Filtering
blur4 = cv.bilateralFilter(img,9,75,75)
plt.subplot(321)
plt.imshow(img)
plt.title('original')
plt.xticks([])
plt.yticks([])
plt.subplot(322)
plt.imshow(blur1)
plt.title('Blurred')
plt.xticks([])
plt.yticks([])
plt.subplot(323)
plt.imshow(blur2)
plt.title('GaussianBlur')
plt.xticks([])
plt.yticks([])
plt.subplot(324)
plt.imshow(blur3)
plt.title('Median Blurring')
plt.xticks([])
plt.yticks([])
plt.subplot(325)
plt.imshow(blur4)
plt.title('Bilate Filter')
plt.xticks([])
plt.yticks([])
plt.show()
```

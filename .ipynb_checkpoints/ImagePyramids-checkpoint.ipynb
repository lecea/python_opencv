```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
img = cv2.imread('qxy.jpg')
edges = cv2.Canny(img,100,200)
lower_reso = cv2.pyrDown(img)
lower_edges = cv2.Canny(lower_reso,100,200)
higher_reso = cv2.pyrUp(img)
higher_edges = cv2.Canny(higher_reso,100,200)
print(img.shape)
print(lower_reso.shape)
print(higher_reso.shape)

plt.subplot(321)
plt.imshow(img,cmap = 'gray')
plt.title('origin')
plt.xticks([])
plt.yticks([])
plt.subplot(322)
plt.imshow(edges,cmap = 'gray')
plt.title('origin')
plt.xticks([])
plt.yticks([])

plt.subplot(323)
plt.imshow(lower_reso,cmap = 'gray')
plt.title('lower_reso')
plt.xticks([])
plt.yticks([])
plt.subplot(324)
plt.imshow(lower_edges,cmap = 'gray')
plt.title('lower_edges')
plt.xticks([])
plt.yticks([])

plt.subplot(325)
plt.imshow(higher_reso,cmap = 'gray')
plt.title('higher_reso')
plt.xticks([])
plt.yticks([])
plt.subplot(326)
plt.imshow(higher_edges,cmap = 'gray')
plt.title('higher_edges')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
import cv2 as cv
import numpy as np,sys
A = cv.imread('qxy.jpg')
A = cv.resize(A, (256,384))
B = cv.imread('qxy.jpg')
B = cv.resize(B, (256,384))
```

```python
# generate Gaussian pyramid for A
G = A.copy()
gpA = [G]
for i in xrange(6):
    G = cv.pyrDown(G)
    gpA.append(G)

# generate Gaussian pyramid for B
G = B.copy()
gpB = [G]
for i in xrange(6):
    G = cv.pyrDown(G)
    gpB.append(G)
```

```python
# generate Laplacian Pyramid for A
lpA = [gpA[5]]
for i in xrange(5,0,-1):
    GE = cv.pyrUp(gpA[i])
    L = cv.subtract(gpA[i-1],GE)
    lpA.append(L)
# generate Laplacian Pyramid for B
lpB = [gpB[5]]
for i in xrange(5,0,-1):
    GE = cv.pyrUp(gpB[i])
    L = cv.subtract(gpB[i-1],GE)
    lpB.append(L)
```

```python
# Now add left and right halves of images in each level
LS = []
for la,lb in zip(lpA,lpB):
    rows,cols,dpt = la.shape
    ls = np.hstack((la[:,0:cols/2],lb[:,cols/2:]))
    LS.append(ls)
# now reconstruct
ls_ = LS[0]
for i in xrange(1,6):
    ls_ = cv.pyrUp(ls_)
    ls_ = cv.add(ls_,LS[i])
# image with direct connecting each half
real = np.hstack((A[:,:cols/2],B[:,cols/2:]))
cv.imwrite('Pyramid_Blending.jpg',ls_)
cv.imwrite('Direct_blending.jpg',real)
```

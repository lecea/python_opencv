```python
import numpy as np
import cv2

# load an color image in grayscale
img = cv2.imread('qxy.jpg')
```

```python
cv2.imshow('image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.namedWindow('image', cv2.WINDOW_NORMAL)
cv2.imshow('image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imwrite('qiuxiyan.png',img)
```

```python
img = cv2.imread('qxy.jpg', 0)
```

```python
cv2.imshow('image', img)
k = cv2.waitKey(0)
if k == 27:
    cv2.destroyAllWindows()
elif k == ord('s'):
    cv2.imwrite('message_qxy.jpg', img)
    cv2.destroyAllWindows()
```

```python
from matplotlib import pyplot as plt

plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([])
plt.yticks([])
plt.show()
```
